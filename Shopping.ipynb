```python
ed = ce.OrdinalEncoder(cols=['age', 'Shop_mode','Lockdown_mode','expensive_online','expensive_offline','buy_online','buy_offline','purchase_online','purches_offline','compare_price_online','compare_price_offline','compare_online_offline','buying_factors','bargin','percent_fraud','prefer_transaction','safe_mode','rate_online','rate_offline','satisfy_online','satisfy_offline'])
df_ed= ed.fit_transform(df)
```

```python
df_ed
```

```python
ax = sns.countplot(y='Shop_mode', data=df);

total = df_ed['Shop_mode'].size
for p in ax.patches:
        percentage = ' {:.1f}%'.format(100 * p.get_width()/total)
        x = p.get_x() + p.get_width() + 0.02
        y = p.get_y() + p.get_height()/2
        ax.annotate(percentage, (x, y))
```

```python
ax = sns.countplot(y='compare_price_offline', data=df_ed);

total = df_ed['compare_price_offline'].size
for p in ax.patches:
        percentage = ' {:.1f}%'.format(100 * p.get_width()/total)
        x = p.get_x() + p.get_width() + 0.02
        y = p.get_y() + p.get_height()/2
        ax.annotate(percentage, (x, y))
```

```python
cat_attributes = df.select_dtypes(include='object')
columns = cat_attributes.columns.tolist()
j = 1

for column in columns:
    p=plt.figure(figsize=(60,30))
    p=plt.subplot(4,5, j)
    
    ax = sns.countplot(y=column, data=cat_attributes)
    
    total = cat_attributes[column].size
    for p in ax.patches:
        percentage = ' {:.1f}%'.format(100 * p.get_width()/total)
        x = p.get_x() + p.get_width() + 0.02
        y = p.get_y() + p.get_height()/2
        ax.annotate(percentage, (x, y))
    
    j += 1
```
